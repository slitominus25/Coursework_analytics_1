```python
### YOUR CODE HERE ###
import datetime
import numpy as np
import pandas as pd

### ПОЛУЧАЕМ ДАТАСЕТЫ ###
courses = pd.read_csv('courses.csv')
course_contents=pd.read_csv('course_contents.csv')
progresses = pd.read_csv('progresses.csv')
progress_phases = pd.read_csv('progress_phases.csv')
students = pd.read_csv('students.csv')
del students['id_']# Удаляем лишний столбец
```

```python
# Создаем список всех возможных дат в период с марта 2016  по июль 2019
date = pd.date_range('2016-03',periods = 41,freq='MS')
#Декартово произведение 
index = pd.MultiIndex.from_product([list_course, date], names = ["course", "date"])
students_count_done_homework = pd.DataFrame(index = index).reset_index()
#Счетчик студентов  
students_count_done_homework['student_count'] = 0
students_count_done_homework
```

```python
# Cоздаем новый DataFrame
final_students_count_done_homework =pd.DataFrame([])
for i in range(len(list_course)):
    name = list_course[i]
    for i,date in students_done_homework.loc[name].iterrows():
            #Создаем маску ,чтобы сортировать по нескольким датам
            mask = (students_count_done_homework['date'] >= date.start_date) & (students_count_done_homework['date'] <= date.finish_date)
            data = students_count_done_homework[students_count_done_homework['course'] == name].loc[mask]
            final_students_count_done_homework = final_students_count_done_homework.append(data)
#Получившийся DataFrame группируем по курсу и дате,все совпадения учитываем в 'student_count'
final_students_count_done_homework =final_students_count_done_homework.groupby(['course','date']).count()
```

```python
list_course_in_final_student_homework = final_student_homework.index.get_level_values('title').unique()
ratio = []
for i in range(len(list_course_in_final_student_homework)):
    name = list_course_in_final_student_homework[i]
    lenght_course = len(final_student_homework.loc[name])
    for module_number,student_count in final_student_homework.loc[name].iterrows():
        if module_number == 1:
            finish_count = student_count[1]
            ratio.append(0)
            continue
        if  module_number == lenght_course:
            start_count = student_count[0]
            ratio.append(start_count/finish_count)
            continue
        start_count = student_count[0]
        ratio.append(start_count/finish_count)
        finish_count = student_count[1]
```
